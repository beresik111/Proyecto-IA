```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, cross_val_score, validation_curve
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, mean_absolute_error
```

```python
#@title Separar X e y
data_df = data_df.dropna()
X = data_df.drop(columns=['id', 'Depression'])
y = data_df['Depression']
```

```python
#@title Preprocesar: One-Hot Encoding para columnas categóricas
categorical_cols = X.select_dtypes(include='object').columns.tolist()
numeric_cols = X.select_dtypes(include=['int64', 'float64']).columns.tolist()

for col in categorical_cols:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col])

X_processed = X.values
```

```python
#@title Split

X_train, X_test, y_train, y_test = train_test_split(X_processed, y, test_size=0.2, random_state=21, shuffle=True)
```

```python
#@title Definir modelos
models = {
    'Decision Tree': DecisionTreeClassifier(random_state=21),
    'Random Forest': RandomForestClassifier(random_state=21),
    'SVM': SVC(random_state=21)
}
```

```python
#@title Entrenar y evaluar
print("\nResultados iniciales:")
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    print(f"{name}: Accuracy = {acc:.4f}")
```

```python
#@title  Función para generar learning curves
def plot_learning_curve(model, param_name, param_range, X, y, title):
    train_scores, test_scores = validation_curve(
        model, X, y, param_name=param_name, param_range=param_range,
        cv=5, scoring='accuracy', n_jobs=-1
    )

    train_mean = np.mean(train_scores, axis=1)
    test_mean = np.mean(test_scores, axis=1)

    plt.figure()
    plt.plot(param_range, train_mean, label='Training score')
    plt.plot(param_range, test_mean, label='Validation score')
    plt.title(title)
    plt.xlabel(param_name)
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)
    plt.show()
```

# Learning Curves

```python
#@title Decision Tree: max_depth
param_range_dt = range(1, 21)
plot_learning_curve(DecisionTreeClassifier(random_state=21), 'max_depth', param_range_dt, X_train, y_train, 'Decision Tree - max_depth')
```

```python
#@title Random Forest: n_estimators
param_range_rf = range(10, 201, 10)
plot_learning_curve(RandomForestClassifier(random_state=21), 'n_estimators', param_range_rf, X_train, y_train, 'Random Forest - n_estimators')
```

```python
#@title SVM: kernel
kernels = ['linear', 'poly', 'rbf']
kernel_scores = []

for kernel in kernels:
    svm = SVC(kernel=kernel, random_state=21)
    scores = cross_val_score(svm, X_train, y_train, cv=5, scoring='accuracy')
    kernel_scores.append(scores.mean())

plt.figure()
plt.bar(kernels, kernel_scores)
plt.title('SVM - Kernel Comparison')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()
```

```python
#@title Cross-validation para Decision Tree

k_values = range(2, 11)
mean_scores = []

for k in k_values:
    scores = cross_val_score(DecisionTreeClassifier(random_state=21), X_train, y_train, cv=k, scoring='accuracy')
    mean_scores.append(scores.mean())

plt.figure()
plt.plot(k_values, mean_scores, marker='o')
plt.title('Decision Tree - Best k-folds')
plt.xlabel('Number of folds (k)')
plt.ylabel('Cross-validated Accuracy')
plt.grid(True)
plt.show()

best_k = k_values[np.argmax(mean_scores)]
print(f"\nMejor k para Decision Tree (Cross-validation): {best_k} folds")
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import BinaryCrossentropy
from tensorflow.keras.metrics import Accuracy
```

```python
input_dim = X_train.shape[1]
n_hidden_layers = [3, 6, 10]
```

```python
results = {}

for n_layers in n_hidden_layers:
    model = Sequential()
    for _ in range(n_layers):
        model.add(Dense(128, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))

    model.compile(
        optimizer=Adam(),
        loss=BinaryCrossentropy(),
        metrics=['accuracy']
    )

    print(f"\nEntrenando red neuronal con {n_layers} capas ocultas...")
    history = model.fit(
        X_train, y_train,
        validation_data=(X_test, y_test),
        epochs=10,
        batch_size=32,
        verbose=1
    )

    loss, acc = model.evaluate(X_test, y_test, verbose=0)
    print(f"Resultado - Capas: {n_layers}, Loss: {loss:.4f}, Accuracy: {acc:.4f}\n")

    results[f"{n_layers}_layers"] = {'history': history, 'loss': loss, 'accuracy': acc}

print("\nResultados finales de todas las configuraciones:")
for layers, metrics in results.items():
    print(f"{layers}: Loss = {metrics['loss']:.4f}, Accuracy = {metrics['accuracy']:.4f}")
```

```python
for config_name, metrics in results.items():
    history = metrics['history']

    plt.figure()
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title(f'{config_name} — Accuracy por Epoch')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, DBSCAN
from sklearn.preprocessing import StandardScaler

# Asumimos que X_processed y y (labels) ya están definidos
X = X_processed
labels = y if isinstance(y, np.ndarray) else y.values

# 1) Escalado
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 2) PCA con 3 componentes
pca3 = PCA(n_components=3)
X_pca3 = pca3.fit_transform(X_scaled)

# 3) K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=21)
km_labels = kmeans.fit_predict(X_pca3)
centroids = kmeans.cluster_centers_

# — K-Means 2D (PC1 vs PC2) —
plt.figure(figsize=(6,4))
scatter_km2 = plt.scatter(
    X_pca3[:, 0], X_pca3[:, 1],
    c=km_labels, cmap='viridis', s=50
)
# Centroides en 2D
plt.scatter(
    centroids[:, 0], centroids[:, 1],
    c='red', marker='X', s=200, label='Centroides'
)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('K-Means (3 PC) — 2D')
plt.legend(*scatter_km2.legend_elements(), title='Cluster')
plt.grid(True)
plt.tight_layout()
plt.show()

# — K-Means 3D —
fig = plt.figure(figsize=(7,6))
ax = fig.add_subplot(111, projection='3d')
scatter_km3 = ax.scatter(
    X_pca3[:, 0], X_pca3[:, 1], X_pca3[:, 2],
    c=km_labels, cmap='viridis', s=50
)
# Centroides en 3D
ax.scatter(
    centroids[:, 0], centroids[:, 1], centroids[:, 2],
    c='red', marker='X', s=200, label='Centroides'
)
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
ax.set_title('K-Means (3 PC) — 3D')
ax.legend(*scatter_km3.legend_elements(), title='Cluster')
plt.tight_layout()
plt.show()

# 4) DBSCAN clustering
dbscan = DBSCAN(min_samples=2)
db_labels = dbscan.fit_predict(X_pca3)

# — DBSCAN 2D —
plt.figure(figsize=(6,4))
scatter_db2 = plt.scatter(
    X_pca3[:, 0], X_pca3[:, 1],
    c=db_labels, cmap='tab10', s=50
)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('DBSCAN (3 PC) — 2D')
plt.legend(*scatter_db2.legend_elements(), title='Cluster')
plt.grid(True)
plt.tight_layout()
plt.show()

# — DBSCAN 3D —
fig = plt.figure(figsize=(7,6))
ax2 = fig.add_subplot(111, projection='3d')
scatter_db3 = ax2.scatter(
    X_pca3[:, 0], X_pca3[:, 1], X_pca3[:, 2],
    c=db_labels, cmap='tab10', s=50
)
ax2.set_xlabel('PC1')
ax2.set_ylabel('PC2')
ax2.set_zlabel('PC3')
ax2.set_title('DBSCAN (3 PC) — 3D')
ax2.legend(*scatter_db3.legend_elements(), title='Cluster')
plt.tight_layout()
plt.show()
```
